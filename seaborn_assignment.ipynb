```python
#importing pandas-->pd ,matplotlib-->plt ,seaborn--->sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
#how to load data form the dataset
healthexp=sns.load_dataset("healthexp")
healthexp
```

```python
# How to find unique values in dataset
healthexp.Year.unique()
```

```python
healthexp.Country.unique()
```

```python
#THERE ARE SO MANY TYPES OF PLOTS IN DATAFRAMES
# HOW TO USE SCATTER PLOT FOR VISULIZE DATA IN DATASET
sns.scatterplot(x="Country",y="Spending_USD",data=healthexp)
plt.title("Scatter Plot")
plt.xlabel("country")
plt.ylabel("Spending_USD")
plt.show()
```

```python
# HOW TO USE LINE PLOT FOR VISULIZE DATA IN DATASET
sns.lineplot(x="Country",y="Spending_USD",data=healthexp,color='brown')
plt.title("Line Plot")
plt.xlabel("country")
plt.ylabel("Spending_USD")
plt.grid()
plt.show()
```

```python
# HOW TO USE BAR PLOT FOR VISULIZE DATA IN DATASET
sns.barplot(x="Country",y="Spending_USD",data=healthexp)
plt.title("Bar Plot")
plt.xlabel("country")
plt.ylabel("Spending_USD")
plt.grid()
plt.show()
```

```python
# HOW TO USE HISTOGRAM PLOT FOR VISUALIZE DATA IN DATASET
sns.histplot(x="Spending_USD",y="Year",data=healthexp, bins=15 ,kde=True,color="blue")
plt.title("Histogram Plot")
plt.xlabel("Spending_USD")
plt.ylabel("year")
plt.grid()
plt.show()
```

```python
# HOW TO USE BOXPLOT FOR VISULIZATION DATA IN DATASET
sns.boxplot(x="Country",y="Spending_USD",data=healthexp , color='blue')
plt.title("Box Plot")
plt.xlabel("country")
plt.ylabel("Spending_USD")
plt.show()
```

```python
# HOW TO USE VIOLINPLOT FOR VISULIZATION DATA IN DATASET
sns.violinplot(x="Country",y="Spending_USD",data=healthexp,color='magenta')
plt.title("Violin Plot")
plt.xlabel("country")
plt.ylabel("Spending_USD")
plt.show()
```

```python
# HOW TO USE PAIRPLOT FOR VISULIZATION DATA IN DATASET
sns.pairplot(healthexp,hue="Country")
plt.title("Pair Plot")
plt.show()
```

```python
# HOW TO USE FACETGRID FOR VISULIZATION DATA IN DATASET
sns.FacetGrid(healthexp ,height=5).map(plt.scatter,"Country","Spending_USD").add_legend()
plt.title("Facet Grid")
plt.show()
```
